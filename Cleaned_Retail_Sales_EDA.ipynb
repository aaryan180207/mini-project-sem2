# Data Normalization and Data Conversion

---
## Part-A: Data Normalization

```python
# 1. Min-Max Normalization on Total Amount
df['TotalAmount_MinMax'] = (df['Total Amount'] - df['Total Amount'].min()) / \
                           (df['Total Amount'].max() - df['Total Amount'].min())
print('\nMin-Max Normalized Total Amount')
print(df[['Customer ID', 'Total Amount', 'TotalAmount_MinMax']].head(10))
```

```python
# 2. Z-score Normalization on Total Amount
df['TotalAmount_Zscore'] = (df['Total Amount'] - df['Total Amount'].mean()) / df['Total Amount'].std()
print('\nZ-score Normalized Total Amount')
print(df[['Customer ID', 'Total Amount', 'TotalAmount_Zscore']].head(10))
```

```python
# 3. Decimal Scaling on Total Amount
df['TotalAmount_Decimal'] = df['Total Amount'] / 10000
print('\nDecimal Scaled Total Amount')
print(df[['Customer ID', 'Total Amount', 'TotalAmount_Decimal']].head(10))
```

```python
# Normalize Multiple Columns (Min-Max)
cols = ['Age', 'Quantity', 'Price per Unit', 'Total Amount']
df_norm = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())
print('\nNormalized Dataset')
print(df_norm.head(10))
```

## NLP Techniques on Text Data

> **Note:** The retail sales dataset does not have a free-text description column like Netflix's `description` field, so full NLP pipelines (tokenisation, POS tagging, word-frequency) are not directly applicable.
> The equivalent analysis here is performed on the **`Product Category`** and **`Gender`** columns to demonstrate the same concepts.


```python
# Token-like analysis — split Product Category values into 'tokens'
from collections import Counter

all_tokens = ' '.join(df['Product Category'].dropna()).split()
token_freq = Counter(all_tokens)
print('Top tokens in Product Category:')
for token, count in token_freq.most_common(10):
    print(f'  {token}: {count}')
```

```python
# Stemming concept - demonstrate with category words
# (NLTK not available in this environment; manual demonstration)
words_demo = ['Electronics', 'Clothing', 'Beauty', 'Beautifully', 'Clothed']
stem_map = {'electronics': 'electron', 'clothing': 'cloth', 'beauty': 'beauti',
            'beautifully': 'beauti', 'clothed': 'cloth'}
for w in words_demo:
    print(w, '->', stem_map.get(w.lower(), w.lower()))
```

# Basic Charts and Visual Encoding

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams.update({'figure.dpi': 120, 'axes.titlesize': 14,
                     'axes.labelsize': 12, 'xtick.labelsize': 10,
                     'ytick.labelsize': 10})

# ── Load & Prepare ───────────────────────────────────────────
df = pd.read_csv('retail_sales_dataset.csv')
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Month'] = df['Date'].dt.to_period('M').astype(str)
df['YearMonth_num'] = df['Date'].dt.year * 100 + df['Date'].dt.month

monthly_df = df.groupby('Month')['Total Amount'].sum().reset_index().sort_values('Month')

gender_cat = df.groupby(['Gender', 'Product Category'])['Total Amount'].sum().unstack(fill_value=0).reset_index()

category_df = df['Product Category'].value_counts().reset_index()
category_df.columns = ['Category', 'Count']

print(df.head())
```

> ## Line Charts and Advanced Line Charts

```python
# 1. Basic Line Chart — Monthly Revenue
fig, ax1 = plt.subplots(figsize=(10, 4))
fig.suptitle('Line Charts', fontsize=16, fontweight='bold', y=1.02)

ax1.plot(monthly_df['Month'], monthly_df['Total Amount'],
         marker='o', color='steelblue', linewidth=2, markersize=7)
ax1.set_title('Monthly Revenue Trend')
ax1.set_xlabel('Month')
ax1.set_ylabel('Total Revenue')
ax1.tick_params(axis='x', rotation=45)
ax1.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# 2. Advanced Line Chart — Revenue by Product Category per Month
monthly_cat = df.groupby(['Month', 'Product Category'])['Total Amount'].sum().unstack(fill_value=0).reset_index().sort_values('Month')

fig, ax2 = plt.subplots(figsize=(10, 5))
fig.suptitle('Comparison Line Charts', fontsize=16, fontweight='bold', y=1.02)

colors_list = ['royalblue', 'tomato', 'mediumseagreen']
markers = ['o', 's', '^']
for i, cat in enumerate(monthly_cat.columns[1:]):
    ax2.plot(monthly_cat['Month'], monthly_cat[cat],
             marker=markers[i % 3], color=colors_list[i % 3],
             linewidth=2, markersize=7, label=cat)
ax2.set_title('Monthly Revenue by Product Category')
ax2.set_xlabel('Month')
ax2.set_ylabel('Revenue')
ax2.legend()
ax2.tick_params(axis='x', rotation=45)
ax2.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout(pad=3.0)
plt.show()
```

```python
# Content Age Distribution Equivalent — Spend Distribution & Yearly Trend
yearly = df.groupby(df['Date'].dt.year)['Total Amount'].sum().reset_index()
yearly.columns = ['Year', 'Total_Revenue']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Spend Distribution & Yearly Revenue', fontsize=16, fontweight='bold', y=1.02)

ax1.hist(df['Total Amount'].dropna(), bins=25,
         color='mediumseagreen', edgecolor='white', linewidth=0.6)
ax1.set_title('Total Amount Spend Distribution')
ax1.set_xlabel('Total Amount')
ax1.set_ylabel('Frequency')
ax1.grid(axis='y', linestyle='--', alpha=0.6)

ax2.fill_between(yearly['Year'], yearly['Total_Revenue'],
                 color='royalblue', alpha=0.5)
ax2.plot(yearly['Year'], yearly['Total_Revenue'],
         color='royalblue', marker='o', linewidth=2)
ax2.set_title('Yearly Revenue Trend')
ax2.set_xlabel('Year')
ax2.set_ylabel('Total Revenue')
ax2.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout(pad=3.0)
plt.show()
```
